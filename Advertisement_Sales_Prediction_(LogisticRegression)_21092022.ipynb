# Advertisement Slae Prediction from an existing customer **LOGISTIC REGRESSION**


Importing Library

```python
import pandas as pd
import numpy as np
```

Load the dataset

```python
dataset = pd.read_csv('Social_Network_Ads_.csv')
print(dataset)
```

Summarize Data

```python
print(dataset.shape) #it shows how many rows and columns are in the dataset
print(dataset.head(5)) #it shows the top five row datasets
```

Segregate the dataset into X & Y
Where, X = input/independent variable
       Y = output/dependent variable 

```python
X = dataset.iloc[:, :-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

Split the dataset to Train & Test or for Cross Validation

```python
from sklearn.model_selection import train_test_split

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_Train = sc.fit_transform(X_Train) # fit() calculate the mean and varience of each data points for every feautures.
X_Test = sc.transform(X_Test) #transform() will convert the data regarding the value of the feature's mean and varience.

print(X_Train)
```

Training

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_Train, Y_Train)
```

Prediction

```python
testAge = int(input("Enter the age: "))
testSalary = int(input("Enter salary: "))

newCustomer = [[testAge, testSalary]]

result = model.predict(sc.transform(newCustomer))

if result == 1:
  print("That customer will buy the producrt.")
else:
  print("That customer will not buy the product.")
```

Prediction for all Test Data

```python
Y_Pred = model.predict(X_Test)
```

Test Result Accuracy

```python
from sklearn.metrics import accuracy_score

print("Accuracy of the model: {0}%".format(accuracy_score(Y_Test, Y_Pred)*100))
```
